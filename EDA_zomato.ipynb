### 1. Analysing Missing Values


```python
df.isnull().sum()
```

```python
# get all missing columns in list
missing_value_features = [feature for feature in df.columns if df[feature].isnull().sum() > 0]
missing_value_features
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
df_country = pd.read_excel('dataset/Zomatodataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df_with_country = pd.merge(df, df_country, on='Country Code', how='left')
df_with_country.head()
```

```python
country_names, country_values = df_with_country.Country.value_counts().index, df_with_country.Country.value_counts().values
```

```python
country_names
```

```python
# matplotlib.rcParams['figure.figsize'] = (18, 6)
plt.pie(country_values[:3], labels=None, autopct='%1.2f%%')
plt.legend(country_names[:3])
# plt.figure(figsize=(8, 8))
# plt.axis('equal')  
```

```python
# Observation - Zomato maximum transactions are from India Then USA then UK
```

```python
df.columns
```

```python
ratings = df_with_country.groupby(['Country','Rating text']).size().reset_index().rename(columns={0:'rating_count'})
ratings

```

```python
excellent_rating = ratings[ratings['Rating text']=='Excellent']
excellent_rating.reset_index(drop=True)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (18, 6)
sns.barplot(data=excellent_rating, x='Country', y='rating_count')
```

```python
df_with_country.columns
```

```python
all_ratings = df_with_country.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'ratings_count'})
all_ratings
```

```python
sns.barplot(all_ratings, x='Aggregate rating', y='ratings_count',hue='Rating text',palette=['Blue', 'Red', 'Orange', 'Yellow', 'Green','Green'])
```

```python
all_ratings['Rating color'].unique()
```

```python
# find the country which has many 0 ratings
zero = df_with_country[df_with_country['Aggregate rating']==0.0].groupby(['Country','Rating color']).size().reset_index().rename(columns={0:'zero_ratings'})
zero.nlargest(1, 'zero_ratings')
```

```python
# India has the most zero ratings inthis transactions
```
